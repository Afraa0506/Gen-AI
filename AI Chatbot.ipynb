```python
import requests

API_KEY = "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"  # Replace with your actual key

def get_available_models(api_key):
    url = "https://openrouter.ai/api/v1/models"
    headers = {
        "Authorization": f"Bearer {api_key}"
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        models = response.json().get("data", [])
        print("✅ Available Models on OpenRouter:")
        for model in models:
            print("-", model["id"])
    else:
        print(f"❌ Failed to fetch models. Status Code: {response.status_code}, Response: {response.text}")

get_available_models(API_KEY)
```

```python
import openai

# Replace with your actual OpenRouter API key
client = openai.OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="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"
)

FREE_MODEL = "mistralai/mistral-7b-instruct"  # Updated free model on OpenRouter

def chat_with_ai():
    print("🤖 AI Chatbot (Free Model) - Type 'exit' to stop chatting.\n")

    chat_history = []

    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            print("👋 Goodbye!")
            break

        # Add user input to chat history
        chat_history.append({"role": "user", "content": user_input})

        try:
            # Get AI response
            response = client.chat.completions.create(
                model=FREE_MODEL,  # Use a valid free model
                messages=chat_history
            )

            # Validate response
            if response and response.choices:
                ai_message = response.choices[0].message.content
                print(f"🤖 AI: {ai_message}")

                # Add AI response to chat history
                chat_history.append({"role": "assistant", "content": ai_message})
            else:
                print("⚠️ No response from AI. Try again later.")

        except Exception as e:
            print(f"❌ Error: {e}")

# Run the chatbot
chat_with_ai()
```
