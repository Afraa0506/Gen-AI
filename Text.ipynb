```python
import gradio as gr
from transformers import pipeline

# Load a pre-trained text classification model
classifier = pipeline("text-classification")

def classify_text(text):
    # Get the classification result from the model
    result = classifier(text)
    return result[0]['label'], result[0]['score']

# Create a Gradio interface
interface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(placeholder="Enter text here..."),
    outputs=["text", "number"],
    title="Text Classification",
    description="Enter some text to classify it into categories."
)

# Launch the interface
interface.launch()
```

```python
import gradio as gr
import spacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

def pos_tagger(sentence):
    """
    Function to process a sentence and return parts of speech tagging.
    """
    doc = nlp(sentence)
    result = [(token.text, token.pos_) for token in doc]

    # Format the output nicely
    formatted_result = "\n".join([f"{word}: {pos}" for word, pos in result])
    return formatted_result

# Create a Gradio interface
iface = gr.Interface(
    fn=pos_tagger,
    inputs=gr.Textbox(lines=2, placeholder="Enter a sentence..."),
    outputs="text",
    title="Parts of Speech Tagger",
    description="Enter a sentence, and this tool will tag each word with its part of speech."
)

# Launch the Gradio app
iface.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load Hugging Face summarization model
summarizer = pipeline("summarization")

def summarize_text(text, num_sentences):
    if not text.strip():
        return "Please enter some text to summarize."

    # Hugging Face models work based on token count, not sentence count
    max_length = num_sentences * 20  # Approximate 20 words per sentence
    summary = summarizer(text, max_length=max_length, min_length=10, do_sample=False)

    return summary[0]['summary_text']

# Gradio interface
iface = gr.Interface(
    fn=summarize_text,
    inputs=[
        gr.Textbox(lines=10, placeholder="Enter long text here..."),
        gr.Slider(1, 10, value=3, step=1, label="Number of Summary Sentences")
    ],
    outputs="text",
    title="AI Text Summarizer",
    description="Enter a large text and select the number of sentences to summarize it into."
)

iface.launch()
```

```python
import gradio as gr
import language_tool_python
from textblob import TextBlob

# Load English grammar checker
tool = language_tool_python.LanguageToolPublicAPI('en-US')

def correct_text(text):
    """
    Fixes spelling mistakes using TextBlob and grammar errors using LanguageTool.
    """
    if not text.strip():
        return "Please enter some text."

    try:
        # Step 1: Fix spelling mistakes
        corrected_spelling = str(TextBlob(text).correct())

        # Step 2: Fix grammar
        corrected_grammar = tool.correct(corrected_spelling)

        # Step 3: Refine grammar again for better accuracy
        final_correction = tool.correct(corrected_grammar)

        return final_correction
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio Interface
iface = gr.Interface(
    fn=correct_text,
    inputs=gr.Textbox(lines=5, placeholder="Enter text with errors..."),
    outputs="text",
    title="Advanced Grammar & Spell Checker",
    description="Fix grammar and spelling errors in your text. Runs multiple correction steps for better accuracy.",
)

iface.launch()
```

```python
import gradio as gr
from keybert import KeyBERT

# Load KeyBERT model
kw_model = KeyBERT()

def extract_keywords(text):
    """
    Extracts important keywords from the given text using KeyBERT.
    """
    if not text.strip():
        return "Please enter some text."

    try:
        # Extract top 10 keywords
        keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 2), stop_words='english', top_n=10)

        # Format keywords nicely
        return "\n".join([kw[0] for kw in keywords])
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio Interface
iface = gr.Interface(
    fn=extract_keywords,
    inputs=gr.Textbox(lines=10, placeholder="Enter a large text..."),
    outputs="text",
    title="Keyword Extraction Without NLTK",
    description="Extracts the most important keywords from a large text using KeyBERT (No NLTK required!).",
)

iface.launch()
```

```python
import gradio as gr
from transformers import pipeline

# Load a text generation model (GPT-2)
text_generator = pipeline("text-generation", model="gpt2")

def generate_text(prompt, max_length, num_return_sequences):
    """
    Generates text based on the input prompt.
    """
    if not prompt.strip():
        return "Please enter a prompt."

    try:
        generated_texts = text_generator(
            prompt,
            max_length=max_length,
            num_return_sequences=num_return_sequences
        )
        return "\n\n".join([text["generated_text"] for text in generated_texts])
    except Exception as e:
        return f"Error: {str(e)}"

# Create Gradio interface
iface = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(lines=2, placeholder="Enter your prompt..."),
        gr.Slider(50, 300, step=10, value=100, label="Max Length"),
        gr.Slider(1, 5, step=1, value=1, label="Number of Outputs"),
    ],
    outputs="text",
    title="AI Text Generator",
    description="Generate text using GPT-2. Enter a prompt and get AI-generated text.",
)

# Launch the app
iface.launch()
```

```python
from transformers import pipeline
import gradio as gr

# Load GPT-2 text generation model
text_generator = pipeline("text-generation", model="gpt2")

def generate_text(prompt, max_length=200, num_return_sequences=1):
    """
    Generates text based on the input prompt with better completion.
    """
    if not prompt.strip():
        return "Please enter a prompt."

    try:
        generated_texts = text_generator(
            prompt,
            max_length=max_length,
            num_return_sequences=num_return_sequences,
            temperature=0.7,  # More creativity
            top_p=0.9,  # Ensures coherence
            do_sample=True  # Enables diverse outputs
        )
        return "\n\n".join([text["generated_text"] for text in generated_texts])
    except Exception as e:
        return f"Error: {str(e)}"

# Create Gradio interface
iface = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(lines=2, placeholder="Enter your prompt..."),
        gr.Slider(100, 500, step=10, value=200, label="Max Length"),  # Increased default max length
        gr.Slider(1, 5, step=1, value=1, label="Number of Outputs"),
    ],
    outputs="text",
    title="AI Text Generator",
    description="Generate text using GPT-2. Enter a prompt and get AI-generated text."
)

# Launch the app
iface.launch()
```

```python
import gradio as gr
import random

def generate_text(subject, verb, obj):
    """Generates a sentence based on user input."""
    if not subject.strip() or not verb.strip() or not obj.strip():
        return "Please enter a subject, verb, and object."

    sentence = f"{subject} {verb} {obj}."
    return sentence

# Gradio Interface
iface = gr.Interface(
    fn=generate_text,
    inputs=[
        gr.Textbox(placeholder="Enter a subject (e.g., The AI, A robot)", label="Subject"),
        gr.Textbox(placeholder="Enter a verb (e.g., creates, analyzes)", label="Verb"),
        gr.Textbox(placeholder="Enter an object (e.g., new technologies, AI models)", label="Object"),
    ],
    outputs="text",
    title="User-Defined Text Generator",
    description="Enter a subject, verb, and object to generate a custom sentence."
)

iface.launch()
```
